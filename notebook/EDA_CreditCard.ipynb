## Data Preprocessing Steps

```python
credit_df['EDUCATION'] = credit_df['EDUCATION'].map({0:4,1:1,2:2,3:3,4:4,5:4,6:4})
credit_df['EDUCATION'].value_counts()
```

**EDUCATION:** (1=graduate school, 2=university, 3=high school, 4=others)

```python
credit_df['MARRIAGE'] = credit_df['MARRIAGE'].map({0:3,1:1,2:2,3:3})
credit_df['MARRIAGE'].value_counts()
```

**MARRIAGE: Marital status** (1=married, 2=single, 3=others)

```python
# Categorical Columns

cat_col = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_1',
                 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
```

```python
def cat_plot(col):
    
    s = credit_df.groupby(col)['default'].mean()
    
    palette = sns.color_palette("bright")
    sns.set_palette(palette, n_colors=9, desat=None, color_codes=False)
    
    plt.subplot(1,2,1)
    sns.countplot(x=col , data=credit_df, order= s.index)
    
    plt.subplot(1,2,2)
    
    wp = {'linewidth':2,'edgecolor':'black'}
    plt.pie(s, labels=s.index, autopct='%0.1f%%',
            radius=1.5,wedgeprops=wp)
```

```python
cat_plot("SEX")
```

```python
cat_plot('MARRIAGE')
```

```python
cat_plot('EDUCATION')
```

```python
cat_plot('PAY_1')
```

* As the repayment status gets delayed by 1 month, the chances of defaulting increases significantly.
